```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
file_path = 'All_Articles_PreProcessed.csv'
df = pd.read_csv(file_path)

# Basic statistics about the dataset
print("Basic Statistics:")
print(df.info())
```

# Title Analysis

```python
# Column for Title length
df['TI_length'] = df['TI'].apply(lambda x: len(str(x).split()))

# Min, Max Title length
min_length_pmid = df.loc[df['TI_length'].idxmin()]['PMID']
max_length_pmid = df.loc[df['TI_length'].idxmax()]['PMID']

article_lengths = df['TI_length']
min_length = article_lengths.min()
max_length = article_lengths.max()
average_length = article_lengths.mean()
median_length = article_lengths.median()
mode_length = article_lengths.mode().values[0]

print(f"Minimum Title Length    : {min_length} (PMID: {min_length_pmid})")
print(f"Maximum Title Length    : {max_length} (PMID: {max_length_pmid})")
print(f"Average Title Length    : {average_length:.2f}")
print(f"Median Title Length     : {median_length}")
print(f"Mode Title Length       : {mode_length}")
```

```python
# Title lengths plot distribution
plt.figure(figsize=(10, 6))
plt.hist(df['TI_length'], bins=50, range=(0, 30), color='skyblue', edgecolor='black')
plt.title('Distribution of Title Lengths')
plt.xlabel('Number of Words in Title')
plt.ylabel('Frequency')
plt.axvline(average_length, color='red', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(median_length, color='green', linestyle='dashed', linewidth=2, label='Median')
plt.axvline(mode_length, color='purple', linestyle='dashed', linewidth=2, label='Mode')
plt.legend()
plt.show()
```

```python
# Common words in Titles using a word cloud
all_titles = ' '.join(df['TI'].astype(str))
wordcloud = WordCloud(width=800, height=400, random_state=21, max_font_size=110, background_color='white').generate(all_titles)
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.title('Most Common Words in Titles')
plt.show()
```
